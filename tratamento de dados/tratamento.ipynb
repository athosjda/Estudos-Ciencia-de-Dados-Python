# Tratamento dos Dados

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statistics as sts
```

```python
# importar dados
dataframe = pd.read_csv('dados/tempo.csv', sep = ';')
```

# AparÃªncia

```python
dataframe.loc[dataframe['Aparencia'] == 'menos', 'Aparencia'] = 'sol'
```

```python
aparencia = dataframe.groupby(['Aparencia']).size()
aparencia.plot.bar()
plt.show()
```

# Temperatura

```python
mediana_temperatura = sts.median(dataframe['Temperatura'])
mediana_temperatura
```

```python
dataframe.loc[(dataframe['Temperatura'] < -130) | (dataframe['Temperatura'] > 130), 'Temperatura'] = mediana_temperatura
```

```python
grafico_temperatura = sns.displot(dataframe['Temperatura'])
plt.show()
```

# Umidade

```python
mediana_umidade = sts.median(dataframe['Umidade'])
mediana_umidade
```

```python
dataframe.loc[(dataframe['Umidade'] < -130) | (dataframe['Umidade'] > 130), 'Umidade'] = mediana_umidade

dataframe.loc[dataframe['Umidade'].isnull(), 'Umidade'] = mediana_umidade
```

```python
grafico_umidade = sns.displot(dataframe['Umidade'])
plt.show()
```

# Vento

```python
dataframe.loc[dataframe['Vento'].isnull(), 'Vento'] = 'FALSO'
```

```python
vento = dataframe.groupby(['Vento']).size()
vento.plot.bar()
plt.show()
```
